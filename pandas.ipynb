```python
numeric_columns = ['Age', 'Physics', 'Chemistry', 'Mathematics', 'Attendance']
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

categorical_columns = ['Name', 'Gender', 'City']
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

print("\nMissing Values after Cleaning:")
print(df.isnull().sum())
```

```python
# VALIDATE DATA TYPES

df['Age'] = df['Age'].astype(int)
df['Physics'] = df['Physics'].astype(int)
df['Chemsitry'] = df['Chemistry'].astype(int)
df['Mathematics'] = df['Mathematics'].astype(int)
df['Attendance'] = df['Attendance'].astype(int)

print("\nData types after cleaning:")
print(df.dtypes)
```

```python
# HANDLE INVALID VALUES

for column in ['Physics', 'Chemistry', 'Mathematics']:
    df.loc[(df[column]<0) | (df[column] > 100), column] = None
    df[column] = df[column].fillna(df[column].mean()).astype(int)

df.loc[(df['Attendance'] < 0) | (df['Attendance'] > 100), 'Attendance'] = None
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].mean())

print("\nCleaned Dataset Summary:")
print(df.describe())
```

```python
# FINAL DATASET

df.to_csv("C:/Users/haris/Downloads/cleaned_student_data.csv", index=False)
print("\nCleaned dataset saved as 'cleaned_student_data.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("Dataset loaded sucessfully")
print("columns in the dataset:", df.columns)

def create_scatter_plot(dataframe, x_col, y_col):
    plt.figure(figsize=(8, 6))
    plt.scatter(dataframe[x_col], dataframe[y_col], color='pink', alpha=0.7)
    plt.title(f'{y_col} vs {x_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
    plt.show(block=True)
```

```python
create_scatter_plot(df, 'Plasma_Temperature_K', 'Neutron_Counts')
create_scatter_plot(df, 'Magnetic_Field_Tesla', 'Radiation_Index')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].scatter(df['Plasma_Temperature_K'], df['Neutron_Counts'], color='blue', alpha=0.7)
axs[0].set_title('Neutron Counts vs Plasma Temperature')
axs[0].set_xlabel('Plasma Temperature (k)')
axs[0].set_ylabel('Neutron Counts')

x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[1].plot(x, y, color='purple')
axs[1].set_title('Sine Wave')
axs[1].set_xlabel('x')
axs[1].set_ylabel('sin(x)')
plt.tight_layout()
plt.show()
```

```python
def create_3d_scatter(dataframe, x_col, y_col, z_col, color_col):
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    
    scatter = ax.scatter(
        dataframe[x_col], dataframe[y_col], dataframe[z_col], 
        c=dataframe[color_col], cmap='viridis', s=50, alpha=0.8
    )
    
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_zlabel(z_col)
    cbar = plt.colorbar(scatter, pad=0.1)
    cbar.set_label(color_col)
    plt.title("3D Scatter Plot")
    plt.show()
create_3d_scatter(df, "Plasma_Temperature_K", "Neutron_Counts", "Magnetic_Field_Tesla", "Radiation_Index")
```
